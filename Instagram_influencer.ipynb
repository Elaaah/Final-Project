```python
df['Avg. Likes']=[str(i) if type(i)!=str else i for i in df['Avg. Likes']]
```

```python
df['Avg. Likes']=[float(i[:-1])*1000000 if i[-1]=='m' else float(i[:-1])*1000 for i in df['Avg. Likes']]
df
```

```python
df['Followers']=[float(i[:-1])*1000000 for i in df['Followers'] if i[-1]=='m']
df
```

```python
df['Posts']=[str(i) if type(i)!=str else i for i in df['Posts']]
df
```

```python
df['Posts']=[float(i[:-1])*1000 if i[-1]=='k' else float(i) for i in df['Posts']]
df
```

```python
df['New Post Avg. Likes']=[str(i) if type(i)!=str else i for i in df['New Post Avg. Likes']]
df
```

```python
df['New Post Avg. Likes']=[float(i[:-1])*1000000 if i[-1]=='m' else float(i[:-1])*1000 if i[-1]=='k' else float(i) for i in df['New Post Avg. Likes']]
df
```

```python
df['Total Likes']=[float(i[:-1])*1000000 if i[-1]=='m' else float(i[:-1])*1000000000 for i in df['Total Likes']]
df
```

```python
df['Channel Info']=df['Channel Info'].apply(str)
import re
def clean_name(s):
    s=re.sub('\n','',s)
    return s


    
```

```python
df['Channel Info']=df['Channel Info'].apply(lambda x:clean_name(x))
df
```

4. Who are the top 10 influencers in the given dataset based on the following features
● Followers

```python
fo=df.nlargest(10,'Followers')
sns.barplot(x=fo['Channel Info'],y=fo['Followers'])
plt.xlabel('Channel Info')
plt.xticks(rotation=90)
plt.ylabel('Followers')
plt.title("Top 10 Influencers in Followers")
plt.show()
```

● Average likes


```python
Avg=df.nlargest(10,'Avg. Likes')
sns.barplot(x=Avg['Channel Info'],y=Avg['Avg. Likes'])
plt.xlabel('Channel Info')
plt.xticks(rotation=90)
plt.ylabel('Avg. Likes')
plt.title('Top 10 Influencers in Average Likes')
plt.show()
```

● Total Likes

```python
lik=df.nlargest(10,'Total Likes')
sns.barplot(x=lik['Channel Info'],y=lik['Total Likes'])
plt.xlabel('Channel Info')
plt.xticks(rotation=90)
plt.ylabel('Total Likes')
plt.title('Top 10 influencers in Total Likes')
plt.show()
```
